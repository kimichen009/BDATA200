<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">B. Key Concepts</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    A <strong>network</strong> (or <strong>graph</strong>) is a mathematical structure used to model pairwise relationships between objects.
  </p>
  <ul style="color:#333; line-height:1.8;">
    <li><strong>Node (Vertex):</strong> An entity in the network &mdash; a person, a bank account, a city, a gene.</li>
    <li><strong>Edge (Link):</strong> A connection between two nodes &mdash; a friendship, a transaction, a road, a co-expression.</li>
  </ul>
  <h3 style="color: #32006E;">Graph Notation</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    A graph is formally written as <strong>G = (V, E)</strong> where:
  </p>
  <ul style="color:#333; line-height:1.8;">
    <li><strong>V</strong>: set of nodes (vertices).</li>
    <li><strong>E</strong>: set of edges (links).</li>
    <li><strong>n = |V|</strong>: number of nodes.</li>
    <li><strong>m = |E|</strong>: number of edges.</li>
  </ul>
  <h3 style="color: #32006E;">Adjacency Matrix</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    An adjacency matrix <strong>A</strong> is an n &times; n matrix where A<sub>ij</sub> = 1 if there is an edge between node i and node j, and 0 otherwise. For undirected graphs, A is symmetric. For weighted graphs, A<sub>ij</sub> holds the edge weight.
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 1.1 &mdash; Building a Simple Graph and Its Adjacency Matrix</h3>
  <p style="color:#555;"><strong>Task:</strong> Create a small friendship network of 5 people, visualize it, and display the adjacency matrix.</p>
</div>

````python
# Example 1.1 — Simple friendship network
G = nx.Graph()
G.add_edges_from([
    ("Alice", "Bob"), ("Alice", "Carol"), ("Bob", "Carol"),
    ("Carol", "Dave"), ("Dave", "Eve")
])

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: network plot
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, ax=axes[0], with_labels=True,
        node_color="#4B2E83", font_color="#E8B923", font_weight="bold",
        node_size=800, edge_color="#B7A6D5", width=2)
axes[0].set_title("Friendship Network", fontweight="bold", color="#4B2E83")

# Right: adjacency matrix
A = nx.to_pandas_adjacency(G, dtype=int)
axes[1].imshow(A.values, cmap="Purples", aspect="equal")
axes[1].set_xticks(range(len(A))); axes[1].set_xticklabels(A.columns, rotation=45)
axes[1].set_yticks(range(len(A))); axes[1].set_yticklabels(A.index)
for i in range(len(A)):
    for j in range(len(A)):
        axes[1].text(j, i, str(A.values[i, j]), ha="center", va="center",
                     color="white" if A.values[i, j] else "gray", fontweight="bold")
axes[1].set_title("Adjacency Matrix", fontweight="bold", color="#4B2E83")
plt.tight_layout()
plt.show()

print("\nAdjacency Matrix:")
print(A)
print(f"\nG = (V, E) with n = |V| = {G.number_of_nodes()}, m = |E| = {G.number_of_edges()}")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 1.2 &mdash; Social Media Network</h3>
  <p style="color:#555;"><strong>Scenario:</strong> Model a Twitter-like follower network. This is a <em>directed</em> network where an edge from A to B means "A follows B".</p>
</div>

````python
# Example 1.2 — Social media follower network
G_social = nx.DiGraph()
G_social.add_edges_from([
    ("@alice", "@bob"), ("@alice", "@carol"), ("@bob", "@carol"),
    ("@carol", "@alice"), ("@dave", "@alice"), ("@dave", "@bob"),
    ("@eve", "@carol"), ("@eve", "@dave"), ("@bob", "@eve"),
])

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
pos = nx.spring_layout(G_social, seed=7)

# Draw network
nx.draw(G_social, pos, ax=axes[0], with_labels=True,
        node_color="#4B2E83", font_color="#E8B923", font_size=9, font_weight="bold",
        node_size=900, edge_color="#B7A6D5", width=1.5,
        arrows=True, arrowsize=15, connectionstyle="arc3,rad=0.1")
axes[0].set_title("Social Media Follower Network (Directed)", fontweight="bold", color="#4B2E83")

# Adjacency matrix
A_social = nx.to_pandas_adjacency(G_social, dtype=int)
axes[1].imshow(A_social.values, cmap="Purples", aspect="equal")
axes[1].set_xticks(range(len(A_social))); axes[1].set_xticklabels(A_social.columns, rotation=45)
axes[1].set_yticks(range(len(A_social))); axes[1].set_yticklabels(A_social.index)
for i in range(len(A_social)):
    for j in range(len(A_social)):
        axes[1].text(j, i, str(A_social.values[i, j]), ha="center", va="center",
                     color="white" if A_social.values[i, j] else "gray", fontweight="bold")
axes[1].set_title("Adjacency Matrix (Directed)", fontweight="bold", color="#4B2E83")
plt.tight_layout()
plt.show()

print(f"Nodes: {G_social.number_of_nodes()}, Edges: {G_social.number_of_edges()}")
print(f"Note: The adjacency matrix is NOT symmetric because the graph is directed.")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 1.3 &mdash; Corruption Network</h3>
  <p style="color:#555;"><strong>Scenario:</strong> Model a corruption investigation where nodes are politicians and companies, and edges represent suspicious money transfers.</p>
</div>

````python
# Example 1.3 — Corruption / suspicious money transfer network
G_corr = nx.DiGraph()
entities = {
    "Politician_A": "politician", "Politician_B": "politician", "Politician_C": "politician",
    "Company_X": "company", "Company_Y": "company", "Company_Z": "company",
    "Shell_1": "shell", "Shell_2": "shell"
}
G_corr.add_nodes_from(entities.keys())

transfers = [
    ("Company_X", "Shell_1", 500000), ("Shell_1", "Politician_A", 200000),
    ("Shell_1", "Politician_B", 150000), ("Company_Y", "Shell_2", 800000),
    ("Shell_2", "Politician_C", 300000), ("Shell_2", "Politician_A", 100000),
    ("Company_Z", "Company_X", 1200000), ("Company_Z", "Shell_2", 400000),
    ("Politician_B", "Company_Y", 50000),
]
for src, tgt, amt in transfers:
    G_corr.add_edge(src, tgt, weight=amt)

color_map = {"politician": "#E53935", "company": "#4B2E83", "shell": "#F57C00"}
node_colors = [color_map[entities[n]] for n in G_corr.nodes()]

fig, ax = plt.subplots(figsize=(10, 7))
pos = nx.spring_layout(G_corr, seed=42, k=1.8)
nx.draw(G_corr, pos, ax=ax, with_labels=True, font_color="black",
        node_color=node_colors, font_size=8, font_weight="bold",
        node_size=1000, edge_color="#999", width=1.5,
        arrows=True, arrowsize=15, connectionstyle="arc3,rad=0.1")

# Add edge labels with amounts
edge_labels = {(u, v): f"${d['weight']:,.0f}" for u, v, d in G_corr.edges(data=True)}
nx.draw_networkx_edge_labels(G_corr, pos, edge_labels, font_size=7, font_color="#333")

patches = [mpatches.Patch(color=c, label=l.title()) for l, c in color_map.items()]
ax.legend(handles=patches, loc="upper left", fontsize=10)
ax.set_title("Corruption Investigation Network", fontweight="bold", color="#4B2E83", fontsize=14)
plt.tight_layout()
plt.show()

print("\nAdjacency Matrix (weighted):")
print(nx.to_pandas_adjacency(G_corr, weight="weight", dtype=int))
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 1.4 &mdash; Bank Transaction Network</h3>
  <p style="color:#555;"><strong>Scenario:</strong> Detect suspicious patterns in a bank transaction network. Edges are weighted by transaction amount.</p>
</div>

````python
# Example 1.4 — Bank transaction network
np.random.seed(42)
accounts = [f"Acct_{i:03d}" for i in range(1, 11)]
G_bank = nx.DiGraph()
G_bank.add_nodes_from(accounts)

transactions = [
    ("Acct_001", "Acct_002", 15000), ("Acct_002", "Acct_003", 14500),
    ("Acct_003", "Acct_004", 14000), ("Acct_004", "Acct_001", 13000),
    ("Acct_005", "Acct_006", 5000),  ("Acct_006", "Acct_007", 3000),
    ("Acct_007", "Acct_008", 8000),  ("Acct_008", "Acct_005", 2000),
    ("Acct_009", "Acct_010", 25000), ("Acct_010", "Acct_001", 20000),
    ("Acct_002", "Acct_005", 7000),  ("Acct_003", "Acct_009", 11000),
]
for s, t, w in transactions:
    G_bank.add_edge(s, t, weight=w)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Network
pos = nx.spring_layout(G_bank, seed=42)
edges = G_bank.edges(data=True)
weights = [d["weight"] / 5000 for _, _, d in edges]
nx.draw(G_bank, pos, ax=axes[0], with_labels=True,
        node_color="#4B2E83", font_color="#E8B923", font_size=8, font_weight="bold",
        node_size=700, edge_color="#B7A6D5", width=weights,
        arrows=True, arrowsize=12)
edge_labels = {(u, v): f"${d['weight']:,.0f}" for u, v, d in G_bank.edges(data=True)}
nx.draw_networkx_edge_labels(G_bank, pos, edge_labels, font_size=6, font_color="#555", ax=axes[0])
axes[0].set_title("Bank Transaction Network", fontweight="bold", color="#4B2E83")

# Adjacency matrix
A_bank = nx.to_pandas_adjacency(G_bank, weight="weight", dtype=int)
im = axes[1].imshow(A_bank.values, cmap="Purples", aspect="equal")
axes[1].set_xticks(range(len(A_bank))); axes[1].set_xticklabels(A_bank.columns, rotation=45, fontsize=7)
axes[1].set_yticks(range(len(A_bank))); axes[1].set_yticklabels(A_bank.index, fontsize=7)
axes[1].set_title("Weighted Adjacency Matrix", fontweight="bold", color="#4B2E83")
plt.colorbar(im, ax=axes[1], shrink=0.8, label="Transaction Amount ($)")
plt.tight_layout()
plt.show()

# Identify circular patterns (potential money laundering)
cycles = list(nx.simple_cycles(G_bank))
print(f"\nCycles detected (potential circular transfers): {len(cycles)}")
for c in cycles[:5]:
    print(f"  {' -> '.join(c)} -> {c[0]}")
````

```
Using NetworkX in Python, create an undirected graph with 8 student nodes and at least 10 edges representing study partnerships. Display the adjacency matrix as a styled heatmap using matplotlib. Find the node with the highest degree. Visualize the network with node sizes proportional to degree. Use purple (#4B2E83) for node colors.
```

````python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# 1. Create an undirected graph with 8 student nodes and at least 10 edges
students = ["Alex", "Ben", "Cathy", "David", "Emily", "Frank", "Grace", "Henry"]
G_study = nx.Graph()
G_study.add_nodes_from(students)

study_partnerships = [
    ("Alex", "Ben"), ("Alex", "Cathy"), ("Alex", "David"),
    ("Ben", "Cathy"), ("Ben", "Emily"),
    ("Cathy", "David"), ("Cathy", "Frank"),
    ("David", "Emily"), ("David", "Grace"),
    ("Emily", "Frank"), ("Emily", "Henry"),
    ("Frank", "Grace")
]
G_study.add_edges_from(study_partnerships)

# 2. Display the adjacency matrix as a heatmap
A_study = nx.to_pandas_adjacency(G_study, dtype=int)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Left: Network Visualization with node sizes proportional to degree
pos = nx.spring_layout(G_study, seed=42)
degrees = dict(G_study.degree())
node_sizes = [degrees[node] * 300 for node in G_study.nodes()] # Scale node size by degree

nx.draw(G_study, pos, ax=axes[0], with_labels=True,
        node_color="#4B2E83", font_color="#E8B923", font_weight="bold",
        node_size=node_sizes, edge_color="#B7A6D5", width=2)
axes[0].set_title("Study Group Network (Node Size = Degree)", fontweight="bold", color="#4B2E83")

# Right: Adjacency Matrix Heatmap
im = axes[1].imshow(A_study.values, cmap="Purples", aspect="equal")
axes[1].set_xticks(range(len(A_study))); axes[1].set_xticklabels(A_study.columns, rotation=45)
axes[1].set_yticks(range(len(A_study))); axes[1].set_yticklabels(A_study.index)
for i in range(len(A_study)):
    for j in range(len(A_study)):
        axes[1].text(j, i, str(A_study.values[i, j]), ha="center", va="center",
                     color="white" if A_study.values[i, j] else "gray", fontweight="bold")
axes[1].set_title("Adjacency Matrix", fontweight="bold", color="#4B2E83")
plt.colorbar(im, ax=axes[1], shrink=0.7)
plt.tight_layout()
plt.show()

# 3. Identify which student has the most study partners
most_partners_student = max(degrees, key=degrees.get)
print(f"\nStudent with the most study partners: {most_partners_student} (Degree: {degrees[most_partners_student]})")

print("\nAdjacency Matrix:")
print(A_study)
````

*Write your interpretation here.*

1. **Which student has the most study partners?** — **Cathy** and **David** both have the highest number of study partners with a degree of 4 each. They are the most central figures in this study group network.
2. **Is the adjacency matrix symmetric? Why?** — Yes, the adjacency matrix is symmetric. This is because the graph is **undirected**, meaning if student A studies with student B, then student B also studies with student A. The connection is mutual, so `A[i,j]` will always equal `A[j,i]`.
3. **What does a large node mean in the visualization?** — A large node in the visualization represents a student with a **higher degree**, meaning they have more direct study partners. Conversely, smaller nodes indicate students with fewer study partners.

````python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# 1. Create a directed graph with 6 employee nodes and weighted edges
employees = ["Alice", "Bob", "Carol", "David", "Eve", "Frank"]
G_email = nx.DiGraph()
G_email.add_nodes_from(employees)

email_exchanges = [
    ("Alice", "Bob", 5), ("Alice", "Carol", 3),
    ("Bob", "David", 8),
    ("Carol", "Bob", 2), ("Carol", "Eve", 6),
    ("David", "Alice", 4),
    ("Eve", "Frank", 7),
    ("Frank", "Carol", 1)
]

for u, v, weight in email_exchanges:
    G_email.add_edge(u, v, weight=weight)

# 2. Display the weighted adjacency matrix
A_email = nx.to_pandas_adjacency(G_email, weight="weight", dtype=int)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Left: Network Visualization with edge width proportional to weight
pos = nx.spring_layout(G_email, seed=42)
edges = G_email.edges(data=True)
widths = [d["weight"] * 0.5 for u, v, d in edges] # Scale edge width by weight

nx.draw(G_email, pos, ax=axes[0], with_labels=True,
        node_color="#4B2E83", font_color="#E8B923", font_weight="bold",
        node_size=1000, edge_color="#B7A6D5", width=widths,
        arrows=True, arrowsize=15, connectionstyle="arc3,rad=0.1")

edge_labels = {(u, v): d['weight'] for u, v, d in G_email.edges(data=True)}
nx.draw_networkx_edge_labels(G_email, pos, edge_labels, font_size=8, font_color="#555", ax=axes[0])
axes[0].set_title("Email Communication Network (Edge Width = Email Count)", fontweight="bold", color="#4B2E83")

# Right: Adjacency Matrix Heatmap
im = axes[1].imshow(A_email.values, cmap="Purples", aspect="equal")
axes[1].set_xticks(range(len(A_email))); axes[1].set_xticklabels(A_email.columns, rotation=45)
axes[1].set_yticks(range(len(A_email))); axes[1].set_yticklabels(A_email.index)
for i in range(len(A_email)):
    for j in range(len(A_email)):
        val = A_email.values[i, j]
        if val > 0:
            axes[1].text(j, i, str(val), ha="center", va="center", color="white", fontweight="bold")
        else:
            axes[1].text(j, i, "0", ha="center", va="center", color="gray", fontweight="normal") # Show 0 for clarity
axes[1].set_title("Weighted Adjacency Matrix", fontweight="bold", color="#4B2E83")
plt.colorbar(im, ax=axes[1], shrink=0.7, label="Email Count")
plt.tight_layout()
plt.show()

# 3. Compute in-degree and out-degree
out_degree = G_email.out_degree(weight='weight')
in_degree = G_email.in_degree(weight='weight')

print("\nEmails Sent (Out-Degree):")
for employee, count in sorted(out_degree, key=lambda item: item[1], reverse=True):
    print(f"  {employee}: {count}")

print("\nEmails Received (In-Degree):")
for employee, count in sorted(in_degree, key=lambda item: item[1], reverse=True):
    print(f"  {employee}: {count}")

# Identify who sends the most and receives the most
most_sender = max(out_degree, key=lambda item: item[1])
most_receiver = max(in_degree, key=lambda item: item[1])

print(f"\nEmployee who sends the most emails: {most_sender[0]} ({most_sender[1]} emails)")
print(f"Employee who receives the most emails: {most_receiver[0]} ({most_receiver[1]} emails)")

print("\nAdjacency Matrix:")
print(A_email)
````

````python
# Example 2.6 — Bipartite: Students and courses
from networkx.algorithms import bipartite

B = nx.Graph()
students = ["Ana", "Ben", "Cara", "Dan", "Eva"]
courses = ["BDATA200", "STAT301", "CS110", "MATH220"]
B.add_nodes_from(students, bipartite=0)
B.add_nodes_from(courses, bipartite=1)
enrollments = [
    ("Ana", "BDATA200"), ("Ana", "STAT301"), ("Ben", "BDATA200"),
    ("Ben", "CS110"), ("Cara", "STAT301"), ("Cara", "MATH220"),
    ("Dan", "CS110"), ("Dan", "MATH220"), ("Dan", "BDATA200"),
    ("Eva", "STAT301"), ("Eva", "CS110"),
]
B.add_edges_from(enrollments)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Bipartite layout
pos = nx.bipartite_layout(B, students)
colors = ["#4B2E83" if n in students else "#F57C00" for n in B.nodes()]
nx.draw(B, pos, ax=axes[0], with_labels=True, font_color="black",
        node_color=colors, font_size=9, font_weight="bold",
        node_size=800, edge_color="#B7A6D5", width=1.5)
patches = [mpatches.Patch(color="#4B2E83", label="Students"),
           mpatches.Patch(color="#F57C00", label="Courses")]
axes[0].legend(handles=patches, loc="upper center")
axes[0].set_title("Bipartite: Student-Course Network", fontweight="bold", color="#4B2E83")

# Projected network: students connected if they share a course
G_proj = bipartite.projected_graph(B, students)
pos2 = nx.spring_layout(G_proj, seed=42)
nx.draw(G_proj, pos2, ax=axes[1], with_labels=True,
        node_color="#4B2E83", font_color="#E8B923", font_weight="bold",
        node_size=800, edge_color="#B7A6D5", width=2)
axes[1].set_title("Projected: Students Sharing Courses", fontweight="bold", color="#4B2E83")
plt.tight_layout()
plt.show()
````

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">C. Density</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Density</strong> measures how connected the entire network is compared with how connected it could possibly be. A dense network has many of the possible connections present, while a sparse network has only a small fraction. Most real-world networks are sparse.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Density is useful because it gives a global summary of the network structure. If density is high, information, influence, or movement may spread more easily. If density is low, the network may be more fragmented or dependent on a few key connections.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    For an undirected graph: <strong>density = 2m / (n(n-1))</strong>, where m is the number of edges and n is the number of nodes.
  </p>
</div>

````python
# Example 3.3 — Comparing density across network types
np.random.seed(42)

fig, axes = plt.subplots(1, 3, figsize=(16, 5))
titles = ["Sparse Network", "Moderate Network", "Dense Network"]

for idx, p in enumerate([0.1, 0.3, 0.7]):
    G_rand = nx.erdos_renyi_graph(15, p, seed=42)
    pos = nx.spring_layout(G_rand, seed=42)
    nx.draw(G_rand, pos, ax=axes[idx], with_labels=True,
            node_color="#4B2E83", font_color="#E8B923", font_weight="bold",
            node_size=400, edge_color="#B7A6D5", width=1.5, font_size=8)
    d = nx.density(G_rand)
    axes[idx].set_title(f"{titles[idx]}\nDensity = {d:.3f}", fontweight="bold", color="#4B2E83")
plt.tight_layout()
plt.show()

# Real-world comparison
print("Density of example networks:")
print(f"  Karate Club: {nx.density(nx.karate_club_graph()):.4f}")
print(f"  Collaboration network: {nx.density(G_collab):.4f}")
print(f"  Bank transaction network: {nx.density(G_bank):.4f}")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">D. Betweenness Centrality</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Betweenness centrality</strong> measures how often a node lies on the shortest paths between other nodes. This is one of the most useful metrics for identifying <strong>bridges, brokers, gatekeepers, and bottlenecks</strong>.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    A node with high betweenness centrality may not have the largest number of connections, but it plays an important role in connecting different parts of the network. In fraud or suspicious-network analysis, high-betweenness nodes may reveal intermediaries connecting different clusters.
  </p>
</div>

````python
# Example 3.4 — Betweenness centrality: Finding bridges
G_bridge = nx.Graph()
# Two clusters connected by a single bridge node
cluster1 = [("A1", "A2"), ("A1", "A3"), ("A2", "A3"), ("A2", "A4"), ("A3", "A4")]
cluster2 = [("B1", "B2"), ("B1", "B3"), ("B2", "B3"), ("B2", "B4"), ("B3", "B4")]
G_bridge.add_edges_from(cluster1 + cluster2)
G_bridge.add_edge("A4", "Bridge")
G_bridge.add_edge("Bridge", "B1")

bc = nx.betweenness_centrality(G_bridge)
degrees = dict(G_bridge.degree())

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
pos = nx.spring_layout(G_bridge, seed=42, k=1.5)

# Color by betweenness
bc_vals = [bc[n] for n in G_bridge.nodes()]
nx.draw(G_bridge, pos, ax=axes[0], with_labels=True, font_color="black",
        node_color=bc_vals, cmap=plt.cm.Purples, node_size=800,
        edge_color="#B7A6D5", width=2, font_weight="bold", font_size=9)
axes[0].set_title("Betweenness Centrality\n(darker = higher)", fontweight="bold", color="#4B2E83")

# Bar chart comparison
nodes = list(G_bridge.nodes())
x = range(len(nodes))
axes[1].bar(x, [bc[n] for n in nodes], color="#4B2E83", alpha=0.7, label="Betweenness")
axes[1].bar(x, [degrees[n]/max(degrees.values()) for n in nodes],
            color="#F57C00", alpha=0.5, label="Degree (normalized)")
axes[1].set_xticks(x); axes[1].set_xticklabels(nodes, rotation=45)
axes[1].legend()
axes[1].set_title("Betweenness vs Degree", fontweight="bold", color="#4B2E83")
plt.tight_layout()
plt.show()

print("Betweenness Centrality:")
for n in sorted(bc, key=bc.get, reverse=True):
    print(f"  {n}: {bc[n]:.4f} (degree={degrees[n]})")
print(f"\nNote: 'Bridge' has the highest betweenness but NOT the highest degree!")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">E. Closeness Centrality</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Closeness centrality</strong> measures how close a node is to all other reachable nodes in the network. A node with high closeness centrality can reach the rest of the network through relatively short paths.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    This metric is useful when we care about <strong>speed of access, information diffusion, communication efficiency, or operational reach</strong>. A node with high closeness is well positioned because it is not far from the rest of the system.
  </p>
</div>

````python
# Example 3.5 — Closeness centrality: Who can reach everyone fastest?
G_kc = nx.karate_club_graph()
cc = nx.closeness_centrality(G_kc)
bc_kc = nx.betweenness_centrality(G_kc)
deg_kc = dict(G_kc.degree())

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
pos = nx.spring_layout(G_kc, seed=42)

# Closeness centrality visualization
cc_vals = [cc[n] for n in G_kc.nodes()]
nx.draw(G_kc, pos, ax=axes[0], with_labels=True, font_color="black",
        node_color=cc_vals, cmap=plt.cm.Purples, node_size=600,
        edge_color="#DDD", width=0.8, font_size=8, font_weight="bold")
axes[0].set_title("Closeness Centrality (darker = closer to all)", fontweight="bold", color="#4B2E83")

# Compare all three centrality measures
metrics_df = pd.DataFrame({
    "Degree": deg_kc, "Betweenness": bc_kc, "Closeness": cc
})
# Normalize for comparison
for col in metrics_df.columns:
    metrics_df[col] = metrics_df[col] / metrics_df[col].max()

top5 = metrics_df.nlargest(5, "Closeness")
top5.plot(kind="bar", ax=axes[1], color=["#4B2E83", "#F57C00", "#1B5E20"], alpha=0.8)
axes[1].set_title("Top 5 by Closeness: All Metrics (normalized)", fontweight="bold", color="#4B2E83")
axes[1].set_ylabel("Normalized Value")
axes[1].legend(loc="upper right")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">F. Clustering Coefficient</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Clustering coefficient</strong> measures how connected a node's neighbors are to each other. If a node has a high clustering coefficient, its neighbors also tend to be connected, forming a tight local group.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    This is useful for identifying <strong>communities, closed groups, or local cohesion</strong>. In a social network, high clustering may mean a close friend group. In a logistics or business network, it may indicate actors that repeatedly interact with one another.
  </p>
</div>

````python
# Example 3.6 — Clustering coefficient: Finding tight groups
G_kc = nx.karate_club_graph()
clustering = nx.clustering(G_kc)
avg_clustering = nx.average_clustering(G_kc)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
pos = nx.spring_layout(G_kc, seed=42)

# Network colored by clustering coefficient
cc_vals = [clustering[n] for n in G_kc.nodes()]
nx.draw(G_kc, pos, ax=axes[0], with_labels=True, font_color="black",
        node_color=cc_vals, cmap=plt.cm.Purples, node_size=600,
        edge_color="#DDD", width=0.8, font_size=8, font_weight="bold")
axes[0].set_title(f"Clustering Coefficient\n(avg = {avg_clustering:.3f})",
                  fontweight="bold", color="#4B2E83")

# Distribution
axes[1].hist(list(clustering.values()), bins=15, color="#4B2E83", edgecolor="white", alpha=0.8)
axes[1].axvline(avg_clustering, color="#F57C00", linestyle="--", linewidth=2, label=f"Average = {avg_clustering:.3f}")
axes[1].set_xlabel("Clustering Coefficient")
axes[1].set_ylabel("Number of Nodes")
axes[1].set_title("Distribution of Clustering Coefficients", fontweight="bold", color="#4B2E83")
axes[1].legend()
plt.tight_layout()
plt.show()

print("\nAll five metrics for the Karate Club graph:")
metrics_summary = pd.DataFrame({
    "Degree": dict(G_kc.degree()),
    "Betweenness": nx.betweenness_centrality(G_kc),
    "Closeness": nx.closeness_centrality(G_kc),
    "Clustering": clustering,
}).round(4)
metrics_summary["Degree"] = metrics_summary["Degree"].astype(int)
print(metrics_summary.to_string())
print(f"\nNetwork Density: {nx.density(G_kc):.4f}")
````

```
Generate three 20-node networks in NetworkX: Erdos-Renyi (p=0.15), Barabasi-Albert (m=2), Watts-Strogatz (k=4, p=0.3). ````python
```
````python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Generate three networks with 20 nodes each
n_nodes = 20
seed = 42

# (a) Erdos-Renyi random (p=0.15)
G_er = nx.erdos_renyi_graph(n_nodes, 0.15, seed=seed)

# (b) Barabasi-Albert preferential attachment (m=2)
G_ba = nx.barabasi_albert_graph(n_nodes, 2, seed=seed)

# (c) Watts-Strogatz small world (k=4, p=0.3)
G_ws = nx.watts_strogatz_graph(n_nodes, 4, 0.3, seed=seed)

networks = {
    "Erdos-Renyi (Random)": G_er,
    "Barabasi-Albert (Scale-free)": G_ba,
    "Watts-Strogatz (Small-world)": G_ws,
}

# Store metrics
metrics_data = []

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
axes = axes.flatten()

for i, (name, G) in enumerate(networks.items()):
    # Compute metrics
    density = nx.density(G)
    avg_clustering = nx.average_clustering(G)
    try:
        avg_shortest_path_length = nx.average_shortest_path_length(G)
    except nx.NetworkXError:
        avg_shortest_path_length = np.nan # Not connected

    metrics_data.append({
        "Network": name,
        "Density": density,
        "Avg Clustering": avg_clustering,
        "Avg Shortest Path Length": avg_shortest_path_length,
    })

    # Visualize the network
    pos = nx.spring_layout(G, seed=seed) # Use a consistent layout algorithm for comparability
    node_colors = plt.cm.Purples(np.array(list(dict(G.degree()).values())) / max(dict(G.degree()).values()))

    nx.draw(G, pos, ax=axes[i], with_labels=True, font_color="black",
            node_color=node_colors, cmap=plt.cm.Purples,
            node_size=800, edge_color="#B7A6D5", width=1.5,
            font_size=8, font_weight="bold")
    axes[i].set_title(f"{name}", fontweight="bold", color="#4B2E83")

plt.suptitle("Comparison of Network Models", fontweight="bold", color="#4B2E83", fontsize=16, y=1.02)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()

metrics_df = pd.DataFrame(metrics_data).set_index("Network")
print("\nComparison of Network Metrics:")
print(metrics_df.round(3))

# 3. Create a comparison bar chart
fig, ax = plt.subplots(figsize=(10, 6))
metrics_df.plot(kind='bar', ax=ax, rot=45, colormap='viridis', alpha=0.8)
ax.set_title("Network Model Metrics Comparison", fontweight="bold", color="#4B2E83")
ax.set_ylabel("Value")
ax.legend(title="Metric")
plt.tight_layout()
plt.show()
````

### Interpretation of Network Models

1.  **Which network is densest?** — The **Erdos-Renyi (Random) graph** is generally the densest. With a fixed probability `p` for every possible edge, it tends to have more connections overall compared to the other models, especially when `p` is not extremely small. The Barabasi-Albert model, though it grows by adding edges, tends to create a sparse network with a few high-degree nodes. Watts-Strogatz also has a relatively low density as it starts with a regular lattice and rewires a few edges.

2.  **Which has highest clustering?** — The **Watts-Strogatz (Small-world) graph** typically exhibits the highest average clustering coefficient. This is due to its construction: it starts as a regular lattice where neighbors of a node are also often neighbors of each other, creating many triangles. Even with rewiring, this high local clustering is largely preserved. Real-world social networks are known for this high clustering, meaning a friend of your friend is likely also your friend.

3.  **Which best resembles a social network and why?** — The **Watts-Strogatz (Small-world) model** generally best resembles a real social network. Social networks are characterized by two key properties: high clustering (people's friends tend to know each other) and small average shortest path length (it doesn't take many steps to get from one person to another). The Watts-Strogatz model successfully captures both of these properties. The Barabasi-Albert model also captures some aspects of social networks (like the presence of hubs, or highly connected individuals), but often with lower clustering than observed in real social graphs. Erdos-Renyi graphs, while having small average shortest path lengths, typically have very low clustering, making them less representative of social structures.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">A. What Is Local Clustering?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Local clustering captures how connected a node's neighbors are to <em>each other</em>. If you know Alice and Bob, and Alice and Bob also know each other, that forms a <strong>triangle</strong> &mdash; a closed relationship.
  </p>
  <h3 style="color: #32006E;">Intuition: Friends of Friends</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    In many real-world networks, a friend of your friend is likely also your friend. This tendency to form closed triangles is called <strong>triadic closure</strong> and is measured by the clustering coefficient.
  </p>
  <h3 style="color: #32006E;">Triangles in a Network</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    A <strong>triangle</strong> is a set of three nodes that are all connected to each other. Triangles are the building blocks of local structure.
  </p>
  <h3 style="color: #32006E;">Local Clustering Coefficient</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    For a node <em>v</em> with degree <em>k</em>, the local clustering coefficient is:
    <br><strong>C(v) = 2T(v) / (k(k-1))</strong>
    <br>where T(v) is the number of triangles through v. It ranges from 0 (no neighbors connected) to 1 (all neighbors connected).
  </p>
  <h3 style="color: #32006E;">Average Clustering Coefficient</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    The average over all nodes: <strong>C&#772; = (1/n) &Sigma; C(v)</strong>. This summarizes how clustered the network is overall.
  </p>
  <h3 style="color: #32006E;">Transitivity</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Transitivity</strong> (global clustering coefficient) = 3 &times; number of triangles / number of connected triples. It gives a single global measure.
  </p>
  <h3 style="color: #32006E;">Interpreting Clustering</h3>
  <ul style="color:#333; line-height:1.8;">
    <li><strong>High clustering:</strong> tight local groups, communities, repeated relationships.</li>
    <li><strong>Low clustering:</strong> tree-like structure, hub-and-spoke, sparse connections.</li>
  </ul>

1.  **Which network has clustering = 1 and why?**
    The **Complete Graph** has an average clustering coefficient and transitivity of 1.0. This is because, by definition, every node in a complete graph is connected to every other node. If a node `A` is connected to `B` and `C`, then `B` and `C` are also connected to each other, forming a perfect triangle. This holds true for all nodes, leading to maximum clustering.

2.  **Which has clustering = 0 and why?**
    The **Star Graph** has an average clustering coefficient and transitivity of 0.0. In a star graph, there is one central node connected to all other peripheral nodes, but the peripheral nodes are *not* connected to each other. For any peripheral node, its only neighbor is the central node, so it cannot form any triangles. For the central node, while it has many neighbors, none of its neighbors are connected to each other, thus it also has a local clustering coefficient of 0. This results in an overall clustering of 0.

3.  **Where does Watts-Strogatz fall and what does this tell us about small-world networks?**
    The **Watts-Strogatz Graph** typically exhibits a moderately high average clustering coefficient (in this example, 0.219). This positions it between the highly clustered complete graph and the unclustered star graph. Watts-Strogatz models are designed to capture the "small-world" phenomenon, which is characterized by both high clustering (like real-world social networks where friends of friends are often friends) and a small average shortest path length. The relatively high clustering, especially compared to random graphs of similar density, is a key feature that makes the Watts-Strogatz model a good representation for many real-world social and biological networks.

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">A. What Does It Mean for a Network to Be Connected?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    A network is <strong>connected</strong> if there is a path between every pair of nodes. If not, it breaks into <strong>connected components</strong> &mdash; separate subgraphs with no edges between them.
  </p>
  <h3 style="color: #32006E;">Types of Components</h3>
  <ul style="color:#333; line-height:1.8;">
    <li><strong>Connected components</strong> (undirected): groups of nodes reachable from each other.</li>
    <li><strong>Weakly connected components</strong> (directed): connected if we ignore edge direction.</li>
    <li><strong>Strongly connected components</strong> (directed): every node can reach every other node following edge direction.</li>
  </ul>
  <h3 style="color: #32006E;">The Giant Component</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    In many real networks, one component contains the majority of nodes. This is called the <strong>giant component</strong>. Understanding its size relative to the whole network tells us about connectivity and fragmentation.
  </p>
  <h3 style="color: #32006E;">Isolates</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Isolates</strong> are nodes with no connections. They form components of size 1 and may represent inactive users, unused facilities, or data errors.
  </p>
</div>

````python
# Example 5.1 — Connected components in an undirected network
G_frag = nx.Graph()
# Three separate groups + isolates
G_frag.add_edges_from([(1,2),(2,3),(3,4),(4,1),(1,3)])  # Component A
G_frag.add_edges_from([(5,6),(6,7),(7,5)])               # Component B
G_frag.add_edges_from([(8,9)])                            # Component C
G_frag.add_nodes_from([10, 11])                           # Isolates

components = list(nx.connected_components(G_frag))
colors_map = ["#4B2E83", "#F57C00", "#1B5E20", "#E53935", "#1565C0"]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
pos = nx.spring_layout(G_frag, seed=42, k=2)

# Color by component
node_colors = []
for node in G_frag.nodes():
    for i, comp in enumerate(components):
        if node in comp:
            node_colors.append(colors_map[i % len(colors_map)])
            break

nx.draw(G_frag, pos, ax=axes[0], with_labels=True, font_color="black",
        node_color=node_colors, font_weight="bold",
        node_size=700, edge_color="#B7A6D5", width=2, font_size=9)
axes[0].set_title(f"Network with {len(components)} Components", fontweight="bold", color="#4B2E83")

# Component size distribution
sizes = sorted([len(c) for c in components], reverse=True)
axes[1].bar(range(len(sizes)), sizes, color="#4B2E83", edgecolor="white")
axes[1].set_xlabel("Component Index")
axes[1].set_ylabel("Number of Nodes")
axes[1].set_title("Component Size Distribution", fontweight="bold", color="#4B2E83")
for i, s in enumerate(sizes):
    axes[1].text(i, s + 0.1, str(s), ha="center", fontweight="bold")
plt.tight_layout()
plt.show()

print(f"Number of components: {len(components)}")
for i, c in enumerate(components):
    print(f"  Component {i+1}: nodes = {sorted(c)}, size = {len(c)}")
print(f"Isolates: {list(nx.isolates(G_frag))}")
print(f"Giant component size: {len(max(components, key=len))} / {G_frag.number_of_nodes()} nodes")
````

````python
# Example 5.2 — Strongly vs weakly connected components (directed)
G_dir = nx.DiGraph()
G_dir.add_edges_from([
    (1,2),(2,3),(3,1),   # Strongly connected cycle
    (3,4),(4,5),         # One-way path
    (6,7),(7,8),(8,6),   # Another strongly connected cycle
    (5,6),               # Bridge
])
G_dir.add_node(9)  # Isolate

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
pos = nx.spring_layout(G_dir, seed=42, k=2)

# Weakly connected components
wcc = list(nx.weakly_connected_components(G_dir))
node_colors_w = []
for node in G_dir.nodes():
    for i, comp in enumerate(wcc):
        if node in comp:
            node_colors_w.append(colors_map[i % len(colors_map)])
            break
nx.draw(G_dir, pos, ax=axes[0], with_labels=True, font_color="black",
        node_color=node_colors_w, font_weight="bold",
        node_size=700, edge_color="#B7A6D5", width=1.5, arrows=True, arrowsize=12)
axes[0].set_title(f"Weakly Connected: {len(wcc)} components", fontweight="bold", color="#4B2E83")

# Strongly connected components
scc = list(nx.strongly_connected_components(G_dir))
node_colors_s = []
for node in G_dir.nodes():
    for i, comp in enumerate(scc):
        if node in comp:
            node_colors_s.append(colors_map[i % len(colors_map)])
            break
nx.draw(G_dir, pos, ax=axes[1], with_labels=True, font_color="black",
        node_color=node_colors_s, font_weight="bold",
        node_size=700, edge_color="#B7A6D5", width=1.5, arrows=True, arrowsize=12)
axes[1].set_title(f"Strongly Connected: {len(scc)} components", fontweight="bold", color="#4B2E83")
plt.tight_layout()
plt.show()

print("Weakly connected components (ignoring direction):")
for i, c in enumerate(wcc):
    print(f"  WCC {i+1}: {sorted(c)}")
print("\nStrongly connected components (following direction):")
for i, c in enumerate(scc):
    print(f"  SCC {i+1}: {sorted(c)}")
````

    *   `{'J'}`
    These are groups of nodes where every node is reachable from every other node within the group *by following the direction of the edges*. Notice that isolated nodes like 'I' and 'J', and nodes at the end of a directed path like 'D' and 'E' (from C->D->E), each form their own SCC if they cannot reach back to other nodes in their WCC.

3.  **Why are there more strongly connected components?** &mdash;
    There are typically more strongly connected components than weakly connected components in a directed graph because the definition of strong connectivity is much stricter. For nodes to be in the same SCC, there must be a directed path *in both directions* between any pair of nodes. If we can travel from A to B but not from B to A (following edge directions), then A and B cannot be in the same SCC, even if they are in the same WCC.

    In this example:
    *   The path `C->D->E` is part of the first WCC, but 'D' and 'E' cannot reach 'C', 'B', or 'A' by following the directed edges. Thus, 'D' and 'E' form separate SCCs.
    *   'I' and 'J' are connected in a WCC (I-J) but only `I->J` exists, not `J->I`. Therefore, they also form separate SCCs.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">A. What Is a Community?</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    A <strong>community</strong> is a group of nodes that are more densely connected to each other than to the rest of the network. Unlike components, communities can overlap and the network can still be fully connected.
  </p>
  <h3 style="color: #32006E;">Components vs Communities</h3>
  <ul style="color:#333; line-height:1.8;">
    <li><strong>Component:</strong> completely disconnected from the rest. No edges cross between components.</li>
    <li><strong>Community:</strong> internally dense, but edges <em>can</em> cross to other communities. Communities exist within connected networks.</li>
  </ul>
  <h3 style="color: #32006E;">Modularity</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Modularity (Q)</strong> measures the quality of a community partition. Higher Q means the division captures real structure. Q ranges from -0.5 to 1.0; values above 0.3 typically indicate significant community structure.
  </p>
  <h3 style="color: #32006E;">Methods</h3>
  <ul style="color:#333; line-height:1.8;">
    <li><strong>Greedy modularity:</strong> iteratively merges communities to maximize Q.</li>
    <li><strong>Louvain:</strong> fast, multi-level optimization &mdash; widely used for large networks.</li>
    <li><strong>Label propagation:</strong> nodes adopt the most common label among their neighbors.</li>
  </ul>
</div>

````python
# Example 6.1 — Community detection on the Karate Club
from community import community_louvain

G_kc = nx.karate_club_graph()

# Three different methods
# 1. Greedy modularity
from networkx.algorithms.community import greedy_modularity_communities
greedy_comms = list(greedy_modularity_communities(G_kc))

# 2. Louvain
louvain_partition = community_louvain.best_partition(G_kc, random_state=42)

# 3. Label propagation
from networkx.algorithms.community import label_propagation_communities
label_comms = list(label_propagation_communities(G_kc))

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
pos = nx.spring_layout(G_kc, seed=42)
comm_colors = ["#4B2E83", "#F57C00", "#1B5E20", "#E53935", "#1565C0", "#6A1B9A"]

# Greedy
colors_g = []
for node in G_kc.nodes():
    for i, comm in enumerate(greedy_comms):
        if node in comm:
            colors_g.append(comm_colors[i % len(comm_colors)])
            break
nx.draw(G_kc, pos, ax=axes[0], with_labels=True, font_color="black", node_color=colors_g,
        node_size=500, edge_color="#DDD", width=0.8, font_size=8, font_weight="bold")
axes[0].set_title(f"Greedy Modularity\n({len(greedy_comms)} communities)", fontweight="bold", color="#4B2E83")

# Louvain
colors_l = [comm_colors[louvain_partition[n] % len(comm_colors)] for n in G_kc.nodes()]
nx.draw(G_kc, pos, ax=axes[1], with_labels=True, font_color="black", node_color=colors_l,
        node_size=500, edge_color="#DDD", width=0.8, font_size=8, font_weight="bold")
n_louvain = len(set(louvain_partition.values()))
axes[1].set_title(f"Louvain\n({n_louvain} communities)", fontweight="bold", color="#4B2E83")

# Label propagation
colors_lp = []
for node in G_kc.nodes():
    for i, comm in enumerate(label_comms):
        if node in comm:
            colors_lp.append(comm_colors[i % len(comm_colors)])
            break
nx.draw(G_kc, pos, ax=axes[2], with_labels=True, font_color="black", node_color=colors_lp,
        node_size=500, edge_color="#DDD", width=0.8, font_size=8, font_weight="bold")
axes[2].set_title(f"Label Propagation\n({len(label_comms)} communities)", fontweight="bold", color="#4B2E83")

plt.tight_layout()
plt.show()

# Modularity comparison
from networkx.algorithms.community import modularity
print("Modularity scores:")
print(f"  Greedy: {modularity(G_kc, greedy_comms):.4f}")
louvain_sets = [set() for _ in range(n_louvain)]
for node, comm_id in louvain_partition.items():
    louvain_sets[comm_id].add(node)
print(f"  Louvain: {modularity(G_kc, louvain_sets):.4f}")
print(f"  Label Propagation: {modularity(G_kc, label_comms):.4f}")
````

````python
# Example 6.2 — Community detection on a larger synthetic network
# Create a network with clear community structure
G_comm = nx.planted_partition_graph(4, 25, 0.3, 0.02, seed=42)

partition = community_louvain.best_partition(G_comm, random_state=42)
n_comms = len(set(partition.values()))

fig, axes = plt.subplots(1, 2, figsize=(16, 7))
pos = nx.spring_layout(G_comm, seed=42, k=0.5)

# Community-colored network
colors = [comm_colors[partition[n] % len(comm_colors)] for n in G_comm.nodes()]
nx.draw(G_comm, pos, ax=axes[0], node_color=colors,
        node_size=100, edge_color="#DDD", width=0.3)
axes[0].set_title(f"Louvain Communities ({n_comms} detected)", fontweight="bold", color="#4B2E83")

# Community sizes
comm_sizes = Counter(partition.values())
axes[1].bar(comm_sizes.keys(), comm_sizes.values(),
            color=[comm_colors[i % len(comm_colors)] for i in comm_sizes.keys()],
            edgecolor="white")
axes[1].set_xlabel("Community ID")
axes[1].set_ylabel("Number of Nodes")
axes[1].set_title("Community Size Distribution", fontweight="bold", color="#4B2E83")
plt.tight_layout()
plt.show()

# Compare central nodes across communities
print(f"\nDetected {n_comms} communities")
bc_comm = nx.betweenness_centrality(G_comm)
for comm_id in sorted(set(partition.values())):
    members = [n for n, c in partition.items() if c == comm_id]
    top_node = max(members, key=lambda n: bc_comm[n])
    print(f"  Community {comm_id}: {len(members)} members, "
          f"top betweenness node = {top_node} ({bc_comm[top_node]:.4f})")
````

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">A. Principles of Effective Network Visualization</h2>
  <ol style="color:#333; line-height:2.2;">
    <li><strong>Start with the question, not the graph.</strong> What are you trying to show?</li>
    <li><strong>Avoid the hairball problem.</strong> A dense graph with all edges looks like a ball of yarn and communicates nothing.</li>
    <li><strong>Choose the right layout.</strong> Spring, circular, shell, spectral, Kamada-Kawai &mdash; each tells a different story.</li>
    <li><strong>Use node size to encode importance</strong> (degree, centrality, revenue).</li>
    <li><strong>Use node color to encode groups</strong> (communities, departments, types).</li>
    <li><strong>Use edge width to encode weight</strong> (transaction amount, frequency).</li>
    <li><strong>Use edge transparency to reduce visual noise.</strong></li>
    <li><strong>Label only the most important nodes.</strong></li>
    <li><strong>Filter large networks before plotting</strong> &mdash; show top-k nodes, largest component, etc.</li>
    <li><strong>Show the largest component when appropriate.</strong></li>
    <li><strong>Compare multiple layouts</strong> before picking one.</li>
    <li><strong>Make directed edges clear with arrows.</strong></li>
    <li><strong>Use legends, titles, and captions.</strong></li>
    <li><strong>Interpret the visualization carefully.</strong> Layout position has no inherent meaning.</li>
  </ol>
</div>

````python
# Example 7.1 — Bad vs Good visualization (same data)
G_large = nx.barabasi_albert_graph(80, 2, seed=42)

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# BAD: hairball, no encoding, all labels
pos_bad = nx.random_layout(G_large, seed=42)
nx.draw(G_large, pos_bad, ax=axes[0], with_labels=True, font_color="black",
        node_color="gray", node_size=200, edge_color="gray",
        width=0.5, font_size=5)
axes[0].set_title("BAD: Random layout, no encoding,\nall labels, uniform style",
                  fontweight="bold", color="#E53935", fontsize=12)

# GOOD: spring layout, size=degree, color=community, top labels, edge alpha
pos_good = nx.spring_layout(G_large, seed=42, k=0.5)
partition = community_louvain.best_partition(G_large, random_state=42)
degrees = dict(G_large.degree())
node_colors = [comm_colors[partition[n] % len(comm_colors)] for n in G_large.nodes()]
node_sizes = [degrees[n] * 50 + 50 for n in G_large.nodes()]

nx.draw_networkx_edges(G_large, pos_good, ax=axes[1], edge_color="#DDD", width=0.3, alpha=0.4)
nx.draw_networkx_nodes(G_large, pos_good, ax=axes[1], node_color=node_colors,
                       node_size=node_sizes, edgecolors="white", linewidths=0.5)

# Label only top 10 nodes by degree
top10 = sorted(degrees, key=degrees.get, reverse=True)[:10]
labels = {n: str(n) for n in top10}
nx.draw_networkx_labels(G_large, pos_good, labels, ax=axes[1], font_size=8, font_weight="bold")

axes[1].set_title("GOOD: Spring layout, size=degree,\ncolor=community, top labels only",
                  fontweight="bold", color="#1B5E20", fontsize=12)
plt.tight_layout()
plt.show()
````

````python
# Example 7.2 — Comparing multiple layouts
G_kc = nx.karate_club_graph()
partition = community_louvain.best_partition(G_kc, random_state=42)
node_colors = [comm_colors[partition[n] % len(comm_colors)] for n in G_kc.nodes()]
degrees = dict(G_kc.degree())
node_sizes = [degrees[n] * 40 + 100 for n in G_kc.nodes()]

layouts = {
    "Spring": nx.spring_layout(G_kc, seed=42),
    "Circular": nx.circular_layout(G_kc),
    "Kamada-Kawai": nx.kamada_kawai_layout(G_kc),
    "Shell": nx.shell_layout(G_kc),
}

fig, axes = plt.subplots(2, 2, figsize=(14, 12))
axes = axes.ravel()

for idx, (name, pos) in enumerate(layouts.items()):
    nx.draw(G_kc, pos, ax=axes[idx], with_labels=True, font_color="black",
            node_color=node_colors, node_size=node_sizes,
            edge_color="#DDD", width=0.8, font_size=8, font_weight="bold")
    axes[idx].set_title(f"{name} Layout", fontweight="bold", color="#4B2E83", fontsize=13)

plt.suptitle("Same Data, Different Layouts", fontweight="bold", color="#4B2E83", fontsize=15, y=1.01)
plt.tight_layout()
plt.show()
````

````python
# Example 7.3 — Creating a clean, publication-ready network visualization
G_kc = nx.karate_club_graph()
partition = community_louvain.best_partition(G_kc, random_state=42)
bc = nx.betweenness_centrality(G_kc)
degrees = dict(G_kc.degree())

fig, ax = plt.subplots(figsize=(12, 9))
pos = nx.spring_layout(G_kc, seed=42, k=0.8)

# Edge transparency based on betweenness of endpoints
for u, v in G_kc.edges():
    avg_bc = (bc[u] + bc[v]) / 2
    alpha = max(0.1, min(avg_bc * 5, 0.8))
    ax.plot([pos[u][0], pos[v][0]], [pos[u][1], pos[v][1]],
            color="#B7A6D5", alpha=alpha, linewidth=0.8)

# Nodes: size=degree, color=community
node_colors = [comm_colors[partition[n] % len(comm_colors)] for n in G_kc.nodes()]
node_sizes = [degrees[n] * 50 + 100 for n in G_kc.nodes()]
nx.draw_networkx_nodes(G_kc, pos, ax=ax, node_color=node_colors,
                       node_size=node_sizes, edgecolors="white", linewidths=1.5)

# Labels only for top betweenness nodes
top_bc = sorted(bc, key=bc.get, reverse=True)[:8]
labels = {n: str(n) for n in top_bc}
nx.draw_networkx_labels(G_kc, pos, labels, ax=ax, font_size=10, font_weight="bold")

# Legend
comm_ids = sorted(set(partition.values()))
legend_patches = [mpatches.Patch(color=comm_colors[i % len(comm_colors)],
                                 label=f"Community {i}") for i in comm_ids]
ax.legend(handles=legend_patches, loc="upper left", fontsize=10, framealpha=0.8)
ax.set_title("Zachary's Karate Club\nNode size = degree | Color = community | Labels = top betweenness",
             fontweight="bold", color="#4B2E83", fontsize=13)
ax.axis("off")
plt.tight_layout()
plt.show()
````

````python
# Example 8.1 — Pyvis: Interactive Karate Club
from pyvis.network import Network

G_kc = nx.karate_club_graph()
partition = community_louvain.best_partition(G_kc, random_state=42)
bc = nx.betweenness_centrality(G_kc)

# Create Pyvis network
net = Network(notebook=True, height="600px", width="100%", bgcolor="#F9F7FC",
              cdn_resources="in_line")

# Add nodes with community colors and size proportional to betweenness
for node in G_kc.nodes():
    color = comm_colors[partition[node] % len(comm_colors)]
    size = 10 + bc[node] * 100
    net.add_node(node, label=str(node), color=color, size=size,
                 title=f"Node {node}\nCommunity: {partition[node]}\nBetweenness: {bc[node]:.3f}")

for u, v in G_kc.edges():
    net.add_edge(u, v, color="#B7A6D5")

net.toggle_physics(True)
net.show("karate_club_interactive.html")
print("Interactive visualization saved to karate_club_interactive.html")
print("Open the file in a browser to explore the network interactively!")
````

````python
# Example 8.2 — Plotly: Interactive network with hover info
import plotly.graph_objects as go

G_kc = nx.karate_club_graph()
pos = nx.spring_layout(G_kc, seed=42)
partition = community_louvain.best_partition(G_kc, random_state=42)
degrees = dict(G_kc.degree())
bc = nx.betweenness_centrality(G_kc)
cc = nx.closeness_centrality(G_kc)

# Edge traces
edge_x, edge_y = [], []
for u, v in G_kc.edges():
    x0, y0 = pos[u]
    x1, y1 = pos[v]
    edge_x.extend([x0, x1, None])
    edge_y.extend([y0, y1, None])

edge_trace = go.Scatter(x=edge_x, y=edge_y, mode="lines",
                        line=dict(width=0.5, color="#B7A6D5"),
                        hoverinfo="none")

# Node traces
node_x = [pos[n][0] for n in G_kc.nodes()]
node_y = [pos[n][1] for n in G_kc.nodes()]
node_colors = [partition[n] for n in G_kc.nodes()]
node_sizes = [degrees[n] * 3 + 8 for n in G_kc.nodes()]
node_text = [f"Node: {n}<br>Degree: {degrees[n]}<br>"
             f"Community: {partition[n]}<br>"
             f"Betweenness: {bc[n]:.3f}<br>"
             f"Closeness: {cc[n]:.3f}"
             for n in G_kc.nodes()]

node_trace = go.Scatter(x=node_x, y=node_y, mode="markers+text",
                        marker=dict(size=node_sizes, color=node_colors,
                                    colorscale="Viridis", line=dict(width=1, color="white")),
                        text=[str(n) for n in G_kc.nodes()],
                        textposition="top center", textfont=dict(size=8),
                        hovertext=node_text, hoverinfo="text")

fig = go.Figure(data=[edge_trace, node_trace],
                layout=go.Layout(
                    title="Karate Club — Interactive Plotly Visualization",
                    titlefont=dict(size=16, color="#4B2E83"),
                    showlegend=False,
                    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    plot_bgcolor="white",
                    hovermode="closest"))
fig.show()
````

````python
# Example 8.3 — Large network: Les Miserables community detection
G_les = nx.les_miserables_graph()

# Compute metrics
partition = community_louvain.best_partition(G_les, random_state=42)
bc = nx.betweenness_centrality(G_les)
degrees = dict(G_les.degree())
cc = nx.clustering(G_les)

n_comms = len(set(partition.values()))
print(f"Les Miserables: {G_les.number_of_nodes()} nodes, {G_les.number_of_edges()} edges")
print(f"Communities detected: {n_comms}")
print(f"Density: {nx.density(G_les):.4f}")
print(f"Average clustering: {nx.average_clustering(G_les):.4f}")

fig, ax = plt.subplots(figsize=(14, 10))
pos = nx.spring_layout(G_les, seed=42, k=0.8)

# Edges with transparency
for u, v, d in G_les.edges(data=True):
    w = d.get("weight", 1)
    alpha = min(0.8, w / 10)
    ax.plot([pos[u][0], pos[v][0]], [pos[u][1], pos[v][1]],
            color="#B7A6D5", alpha=alpha, linewidth=w * 0.3)

# Nodes
extended_colors = plt.cm.tab20(np.linspace(0, 1, n_comms))
node_colors = [extended_colors[partition[n]] for n in G_les.nodes()]
node_sizes = [degrees[n] * 30 + 50 for n in G_les.nodes()]
nx.draw_networkx_nodes(G_les, pos, ax=ax, node_color=node_colors,
                       node_size=node_sizes, edgecolors="white", linewidths=1)

# Label top characters
top_chars = sorted(bc, key=bc.get, reverse=True)[:15]
labels = {n: n for n in top_chars}
nx.draw_networkx_labels(G_les, pos, labels, ax=ax, font_size=8, font_weight="bold")

ax.set_title(f"Les Miserables Character Network\n"
             f"{G_les.number_of_nodes()} characters, {n_comms} communities detected",
             fontweight="bold", color="#4B2E83", fontsize=14)
ax.axis("off")
plt.tight_layout()
plt.show()

# Community analysis
print("\nTop characters by betweenness centrality:")
for n in top_chars[:10]:
    print(f"  {n}: betweenness={bc[n]:.4f}, degree={degrees[n]}, community={partition[n]}")
````

````python
# Example 8.4 — Large scale: Social network with 500 nodes
G_large = nx.powerlaw_cluster_graph(500, 3, 0.5, seed=42)

partition = community_louvain.best_partition(G_large, random_state=42)
n_comms = len(set(partition.values()))
degrees = dict(G_large.degree())

fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Full network
pos = nx.spring_layout(G_large, seed=42, k=0.3, iterations=50)
extended_colors = plt.cm.tab20(np.linspace(0, 1, max(n_comms, 20)))
node_colors = [extended_colors[partition[n] % 20] for n in G_large.nodes()]
node_sizes = [max(degrees[n] * 3, 5) for n in G_large.nodes()]

nx.draw_networkx_edges(G_large, pos, ax=axes[0], edge_color="#DDD", width=0.1, alpha=0.2)
nx.draw_networkx_nodes(G_large, pos, ax=axes[0], node_color=node_colors,
                       node_size=node_sizes, edgecolors="none")
axes[0].set_title(f"500-Node Network ({n_comms} communities)\nFull View",
                  fontweight="bold", color="#4B2E83")
axes[0].axis("off")

# Filtered: top 50 nodes by degree
top50 = sorted(degrees, key=degrees.get, reverse=True)[:50]
G_sub = G_large.subgraph(top50)
pos_sub = nx.spring_layout(G_sub, seed=42)
sub_colors = [extended_colors[partition[n] % 20] for n in G_sub.nodes()]
sub_sizes = [degrees[n] * 10 for n in G_sub.nodes()]
bc_sub = nx.betweenness_centrality(G_sub)

nx.draw(G_sub, pos_sub, ax=axes[1], node_color=sub_colors,
        node_size=sub_sizes, edge_color="#B7A6D5", width=0.5, alpha=0.5)
top5_labels = {n: str(n) for n in sorted(bc_sub, key=bc_sub.get, reverse=True)[:10]}
nx.draw_networkx_labels(G_sub, pos_sub, top5_labels, ax=axes[1],
                        font_size=8, font_weight="bold")
axes[1].set_title("Top 50 Nodes by Degree\nFiltered View", fontweight="bold", color="#4B2E83")
axes[1].axis("off")
plt.tight_layout()
plt.show()

print(f"Full network: {G_large.number_of_nodes()} nodes, {G_large.number_of_edges()} edges")
print(f"Density: {nx.density(G_large):.5f}")
print(f"Communities: {n_comms}")
comm_sizes = Counter(partition.values())
print(f"Largest community: {max(comm_sizes.values())} nodes")
print(f"Smallest community: {min(comm_sizes.values())} nodes")
````

````python
# Example 8.5 — Pyvis: Export Les Miserables as interactive HTML
from pyvis.network import Network

G_les = nx.les_miserables_graph()
partition = community_louvain.best_partition(G_les, random_state=42)
bc = nx.betweenness_centrality(G_les)
degrees = dict(G_les.degree())

net = Network(notebook=True, height="700px", width="100%", bgcolor="#F9F7FC",
              cdn_resources="in_line")

extended_hex = ["#4B2E83", "#F57C00", "#1B5E20", "#E53935", "#1565C0",
                "#6A1B9A", "#00838F", "#BF360C", "#33691E", "#AD1457"]

for node in G_les.nodes():
    color = extended_hex[partition[node] % len(extended_hex)]
    size = 5 + degrees[node] * 2
    title = (f"<b>{node}</b><br>"
             f"Community: {partition[node]}<br>"
             f"Degree: {degrees[node]}<br>"
             f"Betweenness: {bc[node]:.3f}")
    net.add_node(node, label=node, color=color, size=size, title=title)

for u, v, d in G_les.edges(data=True):
    w = d.get("weight", 1)
    net.add_edge(u, v, width=w * 0.5, color="#B7A6D5")

net.toggle_physics(True)
net.show("les_miserables_interactive.html")
print("Saved les_miserables_interactive.html — open in browser to explore!")
````

````python
# Example 8.6 — ipysigma: Efficient large network rendering
# Note: ipysigma works best in Jupyter Notebook/Lab with a running kernel
try:
    from ipysigma import Sigma

    # Build a larger network
    G_power = nx.powerlaw_cluster_graph(300, 3, 0.5, seed=42)
    partition = community_louvain.best_partition(G_power, random_state=42)

    # Add community as node attribute
    nx.set_node_attributes(G_power, partition, "community")

    # Render with ipysigma
    sigma = Sigma(G_power,
                  node_color="community",
                  node_size=G_power.degree,
                  node_label_size=G_power.degree,
                  default_edge_type="curve",
                  max_categorical_colors=20)
    display(sigma)
    print("ipysigma visualization rendered above (interactive in Jupyter)")

except Exception as e:
    print(f"ipysigma not available in this environment: {e}")
    print("To use ipysigma, run this notebook in Jupyter Lab/Notebook.")
    print("\nFalling back to matplotlib...")

    G_power = nx.powerlaw_cluster_graph(300, 3, 0.5, seed=42)
    partition = community_louvain.best_partition(G_power, random_state=42)
    pos = nx.spring_layout(G_power, seed=42, k=0.4)
    colors = [partition[n] for n in G_power.nodes()]
    sizes = [G_power.degree(n) * 5 + 10 for n in G_power.nodes()]

    fig, ax = plt.subplots(figsize=(12, 9))
    nx.draw_networkx_edges(G_power, pos, ax=ax, edge_color="#DDD", width=0.2, alpha=0.2)
    nx.draw_networkx_nodes(G_power, pos, ax=ax, node_color=colors,
                           cmap=plt.cm.tab20, node_size=sizes, edgecolors="none")
    ax.set_title("300-Node Network (matplotlib fallback)", fontweight="bold", color="#4B2E83")
    ax.axis("off")
    plt.tight_layout()
    plt.show()
````
